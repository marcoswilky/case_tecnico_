```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker
```

```python
url = '/content/annual-co2-emissions-per-country.csv'
dados = pd.read_csv(url)
```

```python
dados_brasil = dados[dados['Entity'] == 'Brazil']
dados_brasil = dados_brasil.reset_index(drop=True)
dados_alemanha = dados[dados['Entity'] == 'Germany']
dados_alemanha = dados_alemanha.reset_index(drop=True)
dados_china = dados[dados['Entity'] == 'China']
dados_china = dados_china.reset_index(drop=True)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Comparação de Emissão de CO₂ por Anos', fontsize=16)

dados_brasil['Annual CO₂ emissions Billion Ton'] = dados_brasil['Annual CO₂ emissions'] / 1e9
dados_alemanha['Annual CO₂ emissions Billion Ton'] = dados_alemanha['Annual CO₂ emissions'] / 1e9
dados_china['Annual CO₂ emissions Billion Ton'] = dados_china['Annual CO₂ emissions'] / 1e9

# Concatenate the dataframes for comparison
dados_comparativo = pd.concat([dados_brasil, dados_alemanha, dados_china])
dados_comparativo['Annual CO₂ emissions Billion Ton'] = dados_comparativo['Annual CO₂ emissions'] / 1e9


sns.lineplot(ax=axes[0, 0], data=dados_brasil, x='Year', y='Annual CO₂ emissions Billion Ton', marker='o', markersize=5, color='green')
axes[0, 0].set_title('Brasil')
axes[0, 0].set_xlabel('Ano')
axes[0, 0].set_ylabel('Emissão de CO₂ (Bilhões de Toneladas)')
axes[0, 0].grid(True)
axes[0, 0].yaxis.set_major_formatter(mticker.ScalarFormatter())

sns.lineplot(ax=axes[0, 1], data=dados_alemanha, x='Year', y='Annual CO₂ emissions Billion Ton', marker='o', markersize=5, color='black')
axes[0, 1].set_title('Alemanha')
axes[0, 1].set_xlabel('Ano')
axes[0, 1].set_ylabel('Emissão de CO₂ (Bilhões de Toneladas)')
axes[0, 1].grid(True)
axes[0, 1].yaxis.set_major_formatter(mticker.ScalarFormatter())

sns.lineplot(ax=axes[1, 0], data=dados_china, x='Year', y='Annual CO₂ emissions Billion Ton', marker='o', markersize=5, color='red')
axes[1, 0].set_title('China')
axes[1, 0].set_xlabel('Ano')
axes[1, 0].set_ylabel('Emissão de CO₂ (Bilhões de Toneladas)')
axes[1, 0].grid(True)
axes[1, 0].yaxis.set_major_formatter(mticker.ScalarFormatter())

colors = ['green', 'black', 'red']
sns.lineplot(ax=axes[1, 1], data=dados_comparativo, x='Year', y='Annual CO₂ emissions Billion Ton', hue='Entity', palette=colors, marker='o', markersize=3)
axes[1, 1].set_title('Brasil, Alemanha e China')
axes[1, 1].set_xlabel('Ano')
axes[1, 1].set_ylabel('Emissão de CO₂ (Bilhões de Toneladas)')
axes[1, 1].grid(True)
axes[1, 1].legend(title='País')
axes[1, 1].yaxis.set_major_formatter(mticker.ScalarFormatter())

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Comparação de Emissão de CO₂ por Anos', fontsize=16)

dados_brasil['Annual CO₂ emissions Billion Ton'] = dados_brasil['Annual CO₂ emissions'] / 1e9
dados_alemanha['Annual CO₂ emissions Billion Ton'] = dados_alemanha['Annual CO₂ emissions'] / 1e9
dados_china['Annual CO₂ emissions Billion Ton'] = dados_china['Annual CO₂ emissions'] / 1e9

# Concatenate the dataframes for comparison
dados_comparativo = pd.concat([dados_brasil, dados_alemanha, dados_china])
dados_comparativo['Annual CO₂ emissions Billion Ton'] = dados_comparativo['Annual CO₂ emissions'] / 1e9


sns.lineplot(ax=axes[0, 0], data=dados_brasil, x='Year', y='Annual CO₂ emissions Billion Ton', marker='o', markersize=4, color='green')
axes[0, 0].set_title('Brasil')
axes[0, 0].set_xlabel('Ano')
axes[0, 0].set_ylabel('Emissão de CO₂ (Bilhões de Toneladas)')
axes[0, 0].grid(True)
axes[0, 0].yaxis.set_major_formatter(mticker.ScalarFormatter())
axes[0, 0].fill_between(dados_brasil['Year'], dados_brasil['Annual CO₂ emissions Billion Ton'], color='green', alpha=0.3)


sns.lineplot(ax=axes[0, 1], data=dados_alemanha, x='Year', y='Annual CO₂ emissions Billion Ton', marker='o', markersize=4, color='black')
axes[0, 1].set_title('Alemanha')
axes[0, 1].set_xlabel('Ano')
axes[0, 1].set_ylabel('Emissão de CO₂ (Bilhões de Toneladas)')
axes[0, 1].grid(True)
axes[0, 1].yaxis.set_major_formatter(mticker.ScalarFormatter())
axes[0, 1].fill_between(dados_alemanha['Year'], dados_alemanha['Annual CO₂ emissions Billion Ton'], color='black', alpha=0.3)


sns.lineplot(ax=axes[1, 0], data=dados_china, x='Year', y='Annual CO₂ emissions Billion Ton', marker='o', markersize=4, color='red')
axes[1, 0].set_title('China')
axes[1, 0].set_xlabel('Ano')
axes[1, 0].set_ylabel('Emissão de CO₂ (Bilhões de Toneladas)')
axes[1, 0].grid(True)
axes[1, 0].yaxis.set_major_formatter(mticker.ScalarFormatter())
axes[1, 0].fill_between(dados_china['Year'], dados_china['Annual CO₂ emissions Billion Ton'], color='red', alpha=0.3)


colors = {'Brazil': 'green', 'Germany': 'black', 'China': 'red'}
paises_selecionados = ['Brazil', 'Germany', 'China']
sns.lineplot(ax=axes[1, 1], data=dados_comparativo, x='Year', y='Annual CO₂ emissions Billion Ton', hue='Entity', palette=colors, marker='o', markersize=3)
axes[1, 1].set_title('Brasil, Alemanha e China')
axes[1, 1].set_xlabel('Ano')
axes[1, 1].set_ylabel('Emissão de CO₂ (Bilhões de Toneladas)')
axes[1, 1].grid(True)
axes[1, 1].legend(title='País')
axes[1, 1].yaxis.set_major_formatter(mticker.ScalarFormatter())

# Fill the area under each line in the combined plot
for country in paises_selecionados:
    country_data = dados_comparativo[dados_comparativo['Entity'] == country]
    axes[1, 1].fill_between(country_data['Year'], country_data['Annual CO₂ emissions Billion Ton'], color=colors[country], alpha=0.3)


plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
url_perda_cobertura_arborea = '/content/perde_cobertura_arborea.csv'
perda_cobertura_arborea = pd.read_csv(url_perda_cobertura_arborea)
```

```python
paises_selecionados = ['Brazil', 'China', 'Germany']
dados_filtrados = perda_cobertura_arborea[perda_cobertura_arborea['Entity'].isin(paises_selecionados)].copy()
dados_filtrados['Total tree cover loss Mha'] = dados_filtrados['Total tree cover loss'] / 1e6

plt.figure(figsize=(10, 6))
colors = {'Brazil': 'green', 'China': 'red', 'Germany': 'black'}

# Plot the lines first
sns.lineplot(data=dados_filtrados, x='Year', y='Total tree cover loss Mha', hue='Entity', marker='o', palette=colors)

# Fill the area under each line
for country in paises_selecionados:
    country_data = dados_filtrados[dados_filtrados['Entity'] == country]
    plt.fill_between(country_data['Year'], country_data['Total tree cover loss Mha'], color=colors[country], alpha=0.3)


plt.title('Perda de Cobertura Arbórea ao Longo do Tempo por País')
plt.xlabel('Ano')
plt.ylabel('Perda Total de Cobertura Arbórea (Milhões de Hectares)')
plt.grid(True)

legend_labels = {'Brazil': 'Brasil', 'China': 'China', 'Germany': 'Alemanha'}
handles, labels = plt.gca().get_legend_handles_labels()
translated_labels = [legend_labels[label] for label in labels]
plt.legend(handles, translated_labels, title='País')
plt.show()
```
